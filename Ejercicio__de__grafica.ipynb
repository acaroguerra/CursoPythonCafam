```python
#CAROLINA GUERRA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_excel('/content/composicion_mercado_exp.xlsx')


data = df.pivot_table(index='Empresa', columns='Producto', values='Participación (%)', fill_value=0)


fig, ax = plt.subplots(figsize=(12, 6))
data.plot(kind='bar', stacked=True, ax=ax)


ax.set_title('Composición del Mercado')
ax.set_xlabel('Empresa')
ax.set_ylabel('Participación (%)')
ax.legend(title='Producto')

plt.show()





```
